<h4>Що таке O-нотація і чим вона відрізняється від Ω -нотації?</h4>

O-нотація і Ω-нотація є двома типами математичної нотації, які використовуються для оцінки росту функцій і аналізу складності алгоритмів. Обидві ці нотації належать до сімейства нотацій Асимптотичного зросту.
<h5>O-нотація (Big O-notation):</h5>
<p>1.O-нотація використовується для верхньої межі оцінки асимптотичного поведінки функції.</p>
<p>2.Функція f(n) належить до класу O(g(n)) якщо існує додатна константа c і додатне ціле число n0 ​,такі що для всіх n≥n0, ∣f(n)∣≤c⋅∣g(n)∣.</p>
<h5>Ω-нотація (Big Omega-notation):</h5>
<p>1.
Ω-нотація використовується для нижньої межі оцінки асимптотичного поведінки функції</p>
2.Функція f(n) належить до класу Ω(g(n)) якщо існує додатна константа c і додатне ціле число n0 ​ , такі що для всіх n≥n0 ​ ,∣f(n)∣≥c⋅∣g(n)∣</p>.

<h4>Яку часову складність має алгоритм сортування методом вставляння за
найсприятливіших умов, коли масив відсортовано? Відповідь обґрунтувати.</h4>

<p>При сортуванні методом вставляння вже відсортований масив є ідеальним випадком для цього алгоритму. В такому випадку, кожен елемент масиву вже знаходиться на своєму місці в порядку зростання (або спадання), і алгоритм просто перевіряє кожен елемент і переходить до наступного. Коли алгоритм зустрічає новий елемент, який має більшу (або меншу, залежно від порядку сортування) значеність, він вставляє цей елемент у потрібне місце і продовжує.</p>
<h5>Таким чином, при сортуванні вже відсортованого масиву:</h5>
<p>1.Кожен новий елемент перевіряється для вставки на своє місце.</p>
<p>2.Оскільки елементи вже знаходяться в потрібному порядку, перевірки зазвичай виявляються дуже ефективними.</p>
<p>3.Для кожного елемента потрібно тільки порівняти його з попередніми елементами і вставити його на відповідне місце. Це означає, що кількість порівнянь і переміщень є мінімальними.</p>
Отже, коли масив вже відсортований, алгоритм сортування методом вставки виявляється дуже ефективним і швидким, оскільки він використовує мінімальну кількість операцій порівняння та переміщення елементів.

<h4>Функція часової складності має вигляд: F(N) = N^3 + 7^2 − 14N. Як 
записати асимптотичну складність O-нотації?>і</hї?

Щоб записати асимптотичну складність у O-нотації для функції часової складності F(N)=N^3 +7N^2 −14N, ми шукаємо найвищий ступінь N у цій функції, оскільки в асимптотичних оцінках найбільший член визначає складність.
<h5>Визначення найвищого ступеня N:</h5> 
<p>1.У функції F(N)=N^3 +7N^2−14N, найвищий ступінь N - це N^3 , оскільки це найбільший член за ступенем.</p> 
<p>2.Запис у O-нотації: Отже, асимптотична складність  F(N) у O-нотації буде записана як O(N^3).</p> 
Таким чином, асимптотична складність функції F(N)=N^3 +7N^2−14N у O-нотації буде O(N^3), що означає, що часова складність цієї функції зростає асимптотично як N^3, коли N зростає до нескінченності.

<h4>Реалізувати на Python алгоритм сортування методом бульбашки.</h4>

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        swapped = False

        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        if not swapped:
            break


my_list = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(my_list)
print("Відсортований масив:")
print(my_list)
```
