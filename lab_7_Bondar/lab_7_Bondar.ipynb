<h1>–ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ—ó —Ä–æ–±–æ—Ç–∏:</h1>

<h3>‚àí –°—Ç–≤–æ—Ä–∏—Ç–∏ –±—ñ–Ω–∞—Ä–Ω–µ –¥–µ—Ä–µ–≤–æ –∑–≥—ñ–¥–Ω–æ –∑ –≤–∞—Ä—ñ–∞–Ω—Ç–æ–º, –≤–∏–¥–∞–Ω–∏–º –≤–∏–∫–ª–∞–¥–∞—á–µ–º.</h3>

```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def build_binary_tree():
    # –°—Ç–≤–æ—Ä—é—î–º–æ –≤—É–∑–ª–∏ –¥–ª—è –¥–µ—Ä–µ–≤–∞
    root = TreeNode(1)
    root.left = TreeNode(2)
    root.right = TreeNode(3)
    root.left.left = TreeNode(4)
    root.left.right = TreeNode(5)

    return root

def inorder_traversal(root):
    if root:
        inorder_traversal(root.left)
        print(root.key, end=" ")
        inorder_traversal(root.right)

# –ü–æ–±—É–¥–æ–≤–∞ –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞ –∑–≥—ñ–¥–Ω–æ –∑ –≤–∞—Ä—ñ–∞–Ω—Ç–æ–º
binary_tree = build_binary_tree()

# –í–∏–≤—ñ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É –æ–±—Ö–æ–¥—É –¥–µ—Ä–µ–≤–∞ (inorder traversal)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—Ö–æ–¥—É –¥–µ—Ä–µ–≤–∞ (inorder traversal):")
inorder_traversal(binary_tree)
```

<h3>–ù–∞–ø–∏—Å–∞—Ç–∏ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –≤–∏–¥–∞–ª–µ–Ω–Ω—è –∑–∞–¥–∞–Ω–æ—ó –≥—ñ–ª–∫–∏ –¥–µ—Ä–µ–≤–∞.</h3>

```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.children = []

def delete_branch(root, target_key):
    if not root:
        return None
    
    # –®—É–∫–∞—î–º–æ –¥—ñ—Ç–µ–π, —è–∫—ñ —Ç—Ä–µ–±–∞ –≤–∏–¥–∞–ª–∏—Ç–∏
    root.children = [child for child in root.children if child.key != target_key]
    
    # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ –≤–∏–∫–ª–∏–∫–∞—î–º–æ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –¥–∏—Ç—è—á–æ–≥–æ –≤—É–∑–ª–∞
    for child in root.children:
        delete_branch(child, target_key)

    return root

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è:
def build_sample_tree():
    # –°—Ç–≤–æ—Ä—é—î–º–æ –¥–µ—Ä–µ–≤–æ
    root = TreeNode(1)
    root.children = [TreeNode(2), TreeNode(3), TreeNode(4)]
    root.children[0].children = [TreeNode(5), TreeNode(6)]
    root.children[2].children = [TreeNode(7)]

    return root

def print_tree(root, level=0):
    if root:
        print("  " * level + str(root.key))
        for child in root.children:
            print_tree(child, level + 1)

# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–µ—Ä–µ–≤–æ –¥–ª—è —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è
tree = build_sample_tree()

print("–î–µ—Ä–µ–≤–æ –¥–æ –≤–∏–¥–∞–ª–µ–Ω–Ω—è:")
print_tree(tree)

# –í–∏–¥–∞–ª—è—î–º–æ –≥—ñ–ª–∫—É –∑ –∫–æ—Ä–µ–Ω–µ–≤–∏–º –≤—É–∑–ª–æ–º 3
target_key = 3
delete_branch(tree, target_key)

print("\n–î–µ—Ä–µ–≤–æ –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è –≥—ñ–ª–∫–∏ –∑ –∫–ª—é—á–µ–º", target_key, ":")
print_tree(tree)
```

–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —É –≤–∏–ø–∞–¥–∫—É –∑–±–∞–ª–∞–Ω—Å–æ–≤–∞–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞, –¥–µ –∫–æ–∂–µ–Ω –≤—É–∑–æ–ª –º–∞—î –±–ª–∏–∑—å–∫–æ –ª–æ–≥–∞—Ä–∏—Ñ–º—ñ—á–Ω—É –≥–ª–∏–±–∏–Ω—É.
<h4>–ß–µ—Ä–≤–æ–Ω–æ-—á–æ—Ä–Ω–µ –¥–µ—Ä–µ–≤–æ (Red-Black Tree)</h4>
<h5>Search (–ø–æ—à—É–∫):</h5>
–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ —á–µ—Ä–≤–æ–Ω–æ-—á–æ—Ä–Ω–∏—Ö –¥–µ—Ä–µ–≤.
–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ —á–µ—Ä–≤–æ–Ω–æ-—á–æ—Ä–Ω–∏—Ö –¥–µ—Ä–µ–≤.
<h5>Insert (–≤—Å—Ç–∞–≤–∫–∞):</h5>
–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –ø–æ—Ç—Ä—ñ–±–Ω—ñ—Å—Ç—å –±–∞–ª–∞–Ω—Å—É–≤–∞–Ω–Ω—è –¥–µ—Ä–µ–≤–∞ –ø—ñ—Å–ª—è –≤—Å—Ç–∞–≤–∫–∏.
–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ —á–µ—Ä–≤–æ–Ω–æ-—á–æ—Ä–Ω–∏—Ö –¥–µ—Ä–µ–≤.
<h5>Delete (–≤–∏–¥–∞–ª–µ–Ω–Ω—è):</h5>
–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –ø–æ—Ç—Ä—ñ–±–Ω—ñ—Å—Ç—å –±–∞–ª–∞–Ω—Å—É–≤–∞–Ω–Ω—è –¥–µ—Ä–µ–≤–∞ –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è.
–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ —á–µ—Ä–≤–æ–Ω–æ-—á–æ—Ä–Ω–∏—Ö –¥–µ—Ä–µ–≤.
<h4>AVL-–¥–µ—Ä–µ–≤–æ</h4>

–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ AVL-–¥–µ—Ä–µ–≤–∞.
–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ AVL-–¥–µ—Ä–µ–≤–∞.
<h5>Insert (–≤—Å—Ç–∞–≤–∫–∞):</h5>
–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –ø–æ—Ç—Ä—ñ–±–Ω—ñ—Å—Ç—å –±–∞–ª–∞–Ω—Å—É–≤–∞–Ω–Ω—è –¥–µ—Ä–µ–≤–∞ –ø—ñ—Å–ª—è –≤—Å—Ç–∞–≤–∫–∏.
–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ AVL-–¥–µ—Ä–µ–≤–∞.
<h5>Delete (–≤–∏–¥–∞–ª–µ–Ω–Ω—è):</h5>
–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –ø–æ—Ç—Ä—ñ–±–Ω—ñ—Å—Ç—å –±–∞–ª–∞–Ω—Å—É–≤–∞–Ω–Ω—è –¥–µ—Ä–µ–≤–∞ –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è.
–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫: –û(ùëô–æ–≥(ùëõ)) - —á–µ—Ä–µ–∑ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ AVL-–¥–µ—Ä–µ–≤–∞.

<h3>–ù–∞–ø–∏—Å–∞—Ç–∏ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó –∫—É–ø–∏ –∑ –±—É–¥—å-—è–∫–æ–≥–æ —Ä–∞–Ω–¥–æ–º–Ω–æ–≥–æ –º–∞—Å–∏–≤–∞.</h3>

```python
def heapify(arr):
    n = len(arr)
    
    # –ü–æ–±—É–¥–æ–≤–∞ –∫—É–ø–∏ –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –≤—É–∑–ª–∞ –∑ –Ω–µ–ø—É—Å—Ç–∏–º–∏ –¥—ñ—Ç—å–º–∏
    for i in range(n // 2 - 1, -1, -1):
        heapify_down(arr, i, n)

def heapify_down(arr, i, n):
    smallest = i
    left = 2 * i + 1
    right = 2 * i + 2
    
    # –ü–æ—Ä—ñ–≤–Ω—é—î–º–æ –∑ –ª—ñ–≤–∏–º –¥–∏—Ç–∏–Ω–æ—é
    if left < n and arr[left] < arr[smallest]:
        smallest = left
    
    # –ü–æ—Ä—ñ–≤–Ω—é—î–º–æ –∑ –ø—Ä–∞–≤–∏–º –¥–∏—Ç–∏–Ω–æ—é
    if right < n and arr[right] < arr[smallest]:
        smallest = right
    
    # –Ø–∫—â–æ —ñ–Ω–¥–µ–∫—Å –Ω–∞–π–º–µ–Ω—à–æ–≥–æ –Ω–µ —Å–ø—ñ–≤–ø–∞–¥–∞—î –∑ –ø–æ—Ç–æ—á–Ω–∏–º, –º—ñ–Ω—è—î–º–æ –º—ñ—Å—Ü—è–º–∏
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ –≤–∏–∫–ª–∏–∫–∞—î–º–æ heapify_down –¥–ª—è –Ω–æ–≤–æ—ó –ø–æ–∑–∏—Ü—ñ—ó smallest
        heapify_down(arr, smallest, n)

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
import random

# –ì–µ–Ω–µ—Ä—É—î–º–æ —Ä–∞–Ω–¥–æ–º–Ω–∏–π –º–∞—Å–∏–≤
array = [random.randint(1, 100) for _ in range(10)]
print("Random Array:", array)

# –ë—É–¥—É—î–º–æ –º—ñ–Ω—ñ–º–∞–ª—å–Ω—É –±—ñ–Ω–∞—Ä–Ω—É –∫—É–ø—É –∑ –º–∞—Å–∏–≤—É
heapify(array)
print("Min-Heap:", array)
```

<h3>‚àí –î–æ–¥–∞—Ç–∏ –¥–æ –Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç, —è–∫–∏–π –¥–æ—Ä—ñ–≤–Ω—é—î –≤–∞—à–æ–º—É –ø–æ—Ä—è–¥–∫–æ–≤–æ–º—É –Ω–æ–º–µ—Ä—É —É
—Å–ø–∏—Å–∫—É –≥—Ä—É–ø–∏.
</h3>

```python
def heapify(arr):
    n = len(arr)
    
    # –ü–æ–±—É–¥–æ–≤–∞ –∫—É–ø–∏ –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –≤—É–∑–ª–∞ –∑ –Ω–µ–ø—É—Å—Ç–∏–º–∏ –¥—ñ—Ç—å–º–∏
    for i in range(n // 2 - 1, -1, -1):
        heapify_down(arr, i, n)

def heapify_down(arr, i, n):
    smallest = i
    left = 2 * i + 1
    right = 2 * i + 2
    
    # –ü–æ—Ä—ñ–≤–Ω—é—î–º–æ –∑ –ª—ñ–≤–∏–º –¥–∏—Ç–∏–Ω–æ—é
    if left < n and arr[left] < arr[smallest]:
        smallest = left
    
    # –ü–æ—Ä—ñ–≤–Ω—é—î–º–æ –∑ –ø—Ä–∞–≤–∏–º –¥–∏—Ç–∏–Ω–æ—é
    if right < n and arr[right] < arr[smallest]:
        smallest = right
    
    # –Ø–∫—â–æ —ñ–Ω–¥–µ–∫—Å –Ω–∞–π–º–µ–Ω—à–æ–≥–æ –Ω–µ —Å–ø—ñ–≤–ø–∞–¥–∞—î –∑ –ø–æ—Ç–æ—á–Ω–∏–º, –º—ñ–Ω—è—î–º–æ –º—ñ—Å—Ü—è–º–∏
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ –≤–∏–∫–ª–∏–∫–∞—î–º–æ heapify_down –¥–ª—è –Ω–æ–≤–æ—ó –ø–æ–∑–∏—Ü—ñ—ó smallest
        heapify_down(arr, smallest, n)

def add_to_heap(arr, value):
    arr.append(value)  # –î–æ–¥–∞—î–º–æ –Ω–æ–≤–∏–π –µ–ª–µ–º–µ–Ω—Ç –≤ –∫—ñ–Ω–µ—Ü—å –º–∞—Å–∏–≤—É
    n = len(arr)
    
    # –í—ñ–¥–Ω–æ–≤–ª—é—î–º–æ –≤–ª–∞—Å—Ç–∏–≤—ñ—Å—Ç—å –∫—É–ø–∏
    for i in range(n // 2 - 1, -1, -1):
        heapify_down(arr, i, n)

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
import random

# –ì–µ–Ω–µ—Ä—É—î–º–æ —Ä–∞–Ω–¥–æ–º–Ω–∏–π –º–∞—Å–∏–≤
array = [random.randint(1, 100) for _ in range(10)]
print("Random Array:", array)

# –ë—É–¥—É—î–º–æ –º—ñ–Ω—ñ–º–∞–ª—å–Ω—É –±—ñ–Ω–∞—Ä–Ω—É –∫—É–ø—É –∑ –º–∞—Å–∏–≤—É
heapify(array)
print("Min-Heap:", array)

# –û—Ç—Ä–∏–º—É—î–º–æ –≤–∞—à –ø–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä —É —Å–ø–∏—Å–∫—É –≥—Ä—É–ø–∏
your_group_number = 5  # –ó–∞–º—ñ–Ω—ñ—Ç—å –Ω–∞ —Å–≤—ñ–π –Ω–æ–º–µ—Ä

# –î–æ–¥–∞—î–º–æ –≤–∞—à –ø–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä –¥–æ –∫—É–ø–∏
add_to_heap(array, your_group_number)
print("Min-Heap after adding your group number:", array)
```

<h3>‚àí –í–∏–ª—É—á–∏—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑ –∫—É–ø–∏</h3>

```python
def heapify(arr):
    n = len(arr)
    
    # –ü–æ–±—É–¥–æ–≤–∞ –∫—É–ø–∏ –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –≤—É–∑–ª–∞ –∑ –Ω–µ–ø—É—Å—Ç–∏–º–∏ –¥—ñ—Ç—å–º–∏
    for i in range(n // 2 - 1, -1, -1):
        heapify_down(arr, i, n)

def heapify_down(arr, i, n):
    smallest = i
    left = 2 * i + 1
    right = 2 * i + 2
    
    # –ü–æ—Ä—ñ–≤–Ω—é—î–º–æ –∑ –ª—ñ–≤–∏–º –¥–∏—Ç–∏–Ω–æ—é
    if left < n and arr[left] < arr[smallest]:
        smallest = left
    
    # –ü–æ—Ä—ñ–≤–Ω—é—î–º–æ –∑ –ø—Ä–∞–≤–∏–º –¥–∏—Ç–∏–Ω–æ—é
    if right < n and arr[right] < arr[smallest]:
        smallest = right
    
    # –Ø–∫—â–æ —ñ–Ω–¥–µ–∫—Å –Ω–∞–π–º–µ–Ω—à–æ–≥–æ –Ω–µ —Å–ø—ñ–≤–ø–∞–¥–∞—î –∑ –ø–æ—Ç–æ—á–Ω–∏–º, –º—ñ–Ω—è—î–º–æ –º—ñ—Å—Ü—è–º–∏
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ –≤–∏–∫–ª–∏–∫–∞—î–º–æ heapify_down –¥–ª—è –Ω–æ–≤–æ—ó –ø–æ–∑–∏—Ü—ñ—ó smallest
        heapify_down(arr, smallest, n)

def extract_max_from_heap(arr):
    if len(arr) == 0:
        return None  # –ö—É–ø–∞ –ø–æ—Ä–æ–∂–Ω—è, –Ω–µ–º–∞—î —â–æ –≤–∏–¥–∞–ª—è—Ç–∏
    
    # –ó–∞–º—ñ–Ω–∞ –∫–æ—Ä–µ–Ω—è (–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞) –æ—Å—Ç–∞–Ω–Ω—ñ–º –µ–ª–µ–º–µ–Ω—Ç–æ–º
    max_elem = arr[0]
    arr[0] = arr[-1]
    arr.pop()  # –í–∏–¥–∞–ª—è—î–º–æ –æ—Å—Ç–∞–Ω–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç (–∫–æ–ª–∏—à–Ω—ñ–π –∫–æ—Ä—ñ–Ω—å)
    
    # –í—ñ–¥–Ω–æ–≤–ª—é—î–º–æ –≤–ª–∞—Å—Ç–∏–≤—ñ—Å—Ç—å –∫—É–ø–∏
    heapify_down(arr, 0, len(arr))
    
    return max_elem

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
import random

# –ì–µ–Ω–µ—Ä—É—î–º–æ —Ä–∞–Ω–¥–æ–º–Ω–∏–π –º–∞—Å–∏–≤
array = [random.randint(1, 100) for _ in range(10)]
print("Random Min-Heap Array:", array)

# –ë—É–¥—É—î–º–æ –º—ñ–Ω—ñ–º–∞–ª—å–Ω—É –±—ñ–Ω–∞—Ä–Ω—É –∫—É–ø—É –∑ –º–∞—Å–∏–≤—É
heapify(array)
print("Min-Heap:", array)

# –í–∏–ª—É—á–∞—î–º–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑ –∫—É–ø–∏ (–º—ñ–Ω-–∫—É—á—ñ)
max_element = extract_max_from_heap(array)
print("Max element extracted:", max_element)
print("Min-Heap after extraction:", array)
```

 <h3>–í–∏–≤—á—ñ—Ç—å —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ —Ä—ñ–∑–Ω—ñ –º–µ—Ç–æ–¥–∏ —Ä–æ–∑–≤‚Äô—è–∑–∞–Ω–Ω—è –∫–æ–ª—ñ–∑—ñ–π.</h3>

<h4>1. –í—ñ–¥–∫—Ä–∏—Ç–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Open Addressing):</h4>
<h5>–õ—ñ–Ω—ñ–π–Ω–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Linear Probing):</h5> –ü—Ä–∏ –∑—ñ—Ç–∫–Ω–µ–Ω–Ω—ñ –∫–ª—é—á–∞ –∑ —É–∂–µ —ñ—Å–Ω—É—é—á–∏–º –∫–ª—é—á–µ–º –≤—Å—Ç–∞–≤–∫–∞ –≤—ñ–¥–±—É–≤–∞—î—Ç—å—Å—è –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω—ñ–π –≤—ñ–ª—å–Ω—ñ–π –ø–æ–∑–∏—Ü—ñ—ó —É —Ç–∞–±–ª–∏—Ü—ñ.
<h5>–ö–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Quadratic Probing):</h5> –¶–µ–π –º–µ—Ç–æ–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—ñ –∫—Ä–æ–∫–∏ –¥–ª—è –ø–æ—à—É–∫—É –Ω–∞—Å—Ç—É–ø–Ω–æ—ó –≤—ñ–ª—å–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó —É —Ç–∞–±–ª–∏—Ü—ñ.
<h5>–ü–æ–¥–≤—ñ–π–Ω–µ —Ö–µ—à—É–≤–∞–Ω–Ω—è (Double Hashing):</h5> –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –¥–≤—ñ —Ä—ñ–∑–Ω—ñ —Ö–µ—à-—Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –Ω–æ–≤–∏—Ö –ø–æ–∑–∏—Ü—ñ–π –ø—Ä–∏ –∫–æ–ª—ñ–∑—ñ—è—Ö.
<h4>2. –ó–∞–º–∫–Ω—É—Ç–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Closed Addressing):</h4>
<h5>–õ–∞–Ω—Ü—é–∂–∫–æ–≤–µ –∑–≤'—è–∑—É–≤–∞–Ω–Ω—è (Chaining):</h5> –ö–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç —Ç–∞–±–ª–∏—Ü—ñ —î —Å–ø–∏—Å–∫–æ–º, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç—å –≤—Å—ñ –∑–Ω–∞—á–µ–Ω–Ω—è, —â–æ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—é—Ç—å –æ–¥–Ω–æ–º—É —ñ–Ω–¥–µ–∫—Å—É.
<h5>–ö–ª–∞—Å—Ç–µ—Ä–Ω–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Clustered Probing):</h5> –°–ø—Ä–æ–±–∞ –≤–∏–±–æ—Ä–æ—Ç–∏ –∫–æ–Ω—Ñ–ª—ñ–∫—Ç–∏,

<h3> –†–µ–∞–ª—ñ–∑—É–π—Ç–µ –≥–µ—à-—Ç–∞–±–ª–∏—Ü—é –∑ –ª–∞–Ω—Ü—é–∂–∫–æ–≤–∏–º –≥–µ—à—É–≤–∞–Ω–Ω—è–º.
</h3>

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.table[index] is None:
            # –°—Ç–≤–æ—Ä—é—î–º–æ –Ω–æ–≤–∏–π –≤—É–∑–æ–ª, —è–∫—â–æ –ø–æ–∑–∏—Ü—ñ—è –ø–æ—Ä–æ–∂–Ω—è
            self.table[index] = Node(key, value)
        else:
            # –Ø–∫—â–æ –ø–æ–∑–∏—Ü—ñ—è –Ω–µ –ø–æ—Ä–æ–∂–Ω—è, –¥–æ–¥–∞—î–º–æ –≤ –∫—ñ–Ω–µ—Ü—å –∑–≤'—è–∑–∞–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É
            current = self.table[index]
            while current.next:
                current = current.next
            current.next = Node(key, value)

    def search(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        while current:
            if current.key == key:
                return current.value
            current = current.next
        return None  # –ö–ª—é—á –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ

    def delete(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        prev = None
        while current:
            if current.key == key:
                if prev:
                    prev.next = current.next
                else:
                    self.table[index] = current.next
                return
            prev = current
            current = current.next

    def display(self):
        for i in range(self.size):
            print(f"Bucket {i}: ", end="")
            current = self.table[i]
            while current:
                print(f"({current.key}: {current.value})", end=" -> ")
                current = current.next
            print("None")

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
hash_table = HashTable(10)

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
hash_table.insert("apple", 10)
hash_table.insert("banana", 20)
hash_table.insert("orange", 30)

# –ü–æ—à—É–∫ –µ–ª–µ–º–µ–Ω—Ç–∞
print("Value for 'banana':", hash_table.search("banana"))  # 20

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞
hash_table.delete("banana")

# –í–∏–≤—ñ–¥ –≤—Å—ñ—î—ó —Ç–∞–±–ª–∏—Ü—ñ
hash_table.display()
```

```python
import time
import random
import string
from collections import OrderedDict

class MyHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        if isinstance(key, int):
            return key % self.size
        elif isinstance(key, str):
            return hash(key) % self.size
        elif isinstance(key, list):
            return len(key) % self.size
        elif isinstance(key, dict):
            return len(key) % self.size
        else:
            # For custom objects, assuming they have a unique identifier
            return hash(key.__class__.__name__) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.table[index] is None:
            self.table[index] = [(key, value)]
        else:
            self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        if self.table[index] is not None:
            for k, v in self.table[index]:
                if k == key:
                    return v
        return None

    def delete(self, key):
        index = self.hash_function(key)
        if self.table[index] is not None:
            for i, (k, v) in enumerate(self.table[index]):
                if k == key:
                    del self.table[index][i]
                    break

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è
def test_hash_table(data_type, data_size, num_operations):
    hash_table = MyHashTable(data_size)
    data = generate_data(data_type, data_size)

    # –í—Å—Ç–∞–≤–∫–∞ –¥–∞–Ω–∏—Ö —É –≥–µ—à-—Ç–∞–±–ª–∏—Ü—é
    start_time = time.time()
    for key, value in data:
        hash_table.insert(key, value)
    insert_time = time.time() - start_time

    # –ü–æ—à—É–∫ –¥–∞–Ω–∏—Ö —É –≥–µ—à-—Ç–∞–±–ª–∏—Ü—ñ
    start_time = time.time()
    for key, _ in data:
        hash_table.search(key)
    search_time = time.time() - start_time

    # –í–∏–¥–∞–ª–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑ –≥–µ—à-—Ç–∞–±–ª–∏—Ü—ñ
    start_time = time.time()
    for key, _ in data:
        hash_table.delete(key)
    delete_time = time.time() - start_time

    return insert_time, search_time, delete_time

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó —Ç–µ—Å—Ç–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö
def generate_data(data_type, size):
    data = []
    if data_type == 'int':
        data = [(random.randint(1, 10000), i) for i in range(size)]
    elif data_type == 'str':
        data = [(random_string(), i) for i in range(size)]
    elif data_type == 'list':
        data = [([random.randint(1, 100) for _ in range(random.randint(1, 5))], i) for i in range(size)]
    elif data_type == 'dict':
        data = [({random_string(): random.randint(1, 100) for _ in range(random.randint(1, 3))}, i) for i in range(size)]
    elif data_type == 'object':
        data = [(CustomObject(i), i) for i in range(size)]
    return data

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó –≤–∏–ø–∞–¥–∫–æ–≤–æ–≥–æ —Ä—è–¥–∫–∞
def random_string(length=10):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for _ in range(length))

# –ü—Ä–∏–∫–ª–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤–ª–∞—Å–Ω–æ–≥–æ –æ–±'—î–∫—Ç—É
class CustomObject:
    def __init__(self, id):
        self.id = id

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è –∑ —Ä—ñ–∑–Ω–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö
data_types = ['int', 'str', 'list', 'dict', 'object']
data_size = 1000
num_operations = 1000

for data_type in data_types:
    insert_time, search_time, delete_time = test_hash_table(data_type, data_size, num_operations)
    print(f"Data Type: {data_type}")
    print(f"Insert Time: {insert_time:.6f} seconds")
    print(f"Search Time: {search_time:.6f} seconds")
    print(f"Delete Time: {delete_time:.6f} seconds")
    print()
```

<h3>–ù–∞–≤–µ–¥—ñ—Ç—å –ø—Ä–∏–∫–ª–∞–¥–∏ –∑–∞–¥–∞—á, —è–∫—ñ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –≤–∏—Ä—ñ—à—É—é—Ç—å—Å—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é

<h5>–ü–æ—à—É–∫ –µ–ª–µ–º–µ–Ω—Ç–∞ –∑–∞ –∫–ª—é—á–µ–º (Binary Search Tree):</h5>

<h5>–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö (Binary Search Tree –∞–±–æ Heap):</h5>

<h5>–û—Ä–≥–∞–Ω—ñ–∑–∞—Ü—ñ—è —Ñ–∞–π–ª–æ–≤–æ—ó —Å–∏—Å—Ç–µ–º–∏ (B-–¥–µ—Ä–µ–≤–æ):</h5>

<p>–õ—ñ–Ω—ñ–π–Ω–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Linear Probing): –ü–æ—à—É–∫ –≤—ñ–ª—å–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ–≥–æ –ø–µ—Ä–µ–≥–ª—è–¥—É.</p>
<p>–ö–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–µ –∑–æ–Ω–¥—É–≤–∞–Ω–Ω—è (Quadratic Probing): –ü–æ—à—É–∫ –≤—ñ–ª—å–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ—ó –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ.</p>
<p>–ü–æ–¥–≤—ñ–π–Ω–µ —Ö–µ—à—É–≤–∞–Ω–Ω—è (Double Hashing): –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –¥—Ä—É–≥–æ—ó —Ö–µ—à-—Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –Ω–æ–≤–æ—ó –ø–æ–∑–∏—Ü—ñ—ó –≤ —Ä–∞–∑—ñ –∫–æ–ª—ñ–∑—ñ—ó.</p>
<h5>–ü–µ—Ä–µ–≤–∞–≥–∏:</h5>
<p>–ù–µ –ø–æ—Ç—Ä–µ–±—É—î –¥–æ–¥–∞—Ç–∫–æ–≤–æ–≥–æ –ø—Ä–æ—Å—Ç–æ—Ä—É –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è –≤–∫–∞–∑—ñ–≤–Ω–∏–∫—ñ–≤.</p>
<p>–ú–æ–∂–µ –±—É—Ç–∏ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–∏–º –¥–ª—è –º–∞–ª–∏—Ö —Ç–∞–±–ª–∏—Ü—å –∞–±–æ –ø—Ä–∏ –Ω–∏–∑—å–∫–æ–º—É —Ä—ñ–≤–Ω—ñ –∫–æ–ª—ñ–∑—ñ–π.</p>
<h5>–ù–µ–¥–æ–ª—ñ–∫–∏:</h5>
<p>–ü—Ä–∏ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—ñ —Ñ–∞–∫—Ç–æ—Ä–∞ –∑–∞–ø–æ–≤–Ω–µ–Ω–Ω—è (load factor) —à–∞–Ω—Å–∏ –Ω–∞ –∫–æ–ª—ñ–∑—ñ—ó –∑—Ä–æ—Å—Ç–∞—é—Ç—å.</p>
<p>–ú–æ–∂–µ –≤–∏–Ω–∏–∫–Ω—É—Ç–∏ –ø—Ä–æ–±–ª–µ–º–∞ –∑–∞–º–∫–Ω–µ–Ω–æ—Å—Ç—ñ (clustering), –∫–æ–ª–∏ –≤—ñ–ª—å–Ω—ñ –ø–æ–∑–∏—Ü—ñ—ó —É—Ç–≤–æ—Ä—é—é—Ç—å –≥—Ä—É–ø–∏.</p>
