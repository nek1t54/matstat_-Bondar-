<h3>-оцінити асимптотичну складність алгоритму бінарного пошуку в
О-нотації.</h3>

<p>Основна ідея бінарного пошуку полягає у поділі відсортованого масиву навпіл і подальшому порівнянні шуканого елемента з серединою масиву. Якщо шуканий елемент менший за середину, пошук продовжується в першій половині масиву; якщо більший - в другій половині. Цей процес продовжується, доки не буде знайдений елемент або не буде визначено, що такого елемента немає.</p>

<p>Асимптотична складність O(logn) визначається тим, що кожен крок бінарного пошуку зменшує область пошуку (кількість елементів для подальшого розгляду) приблизно вдвічі. Це відбувається через те, що на кожній ітерації пошуку ми розділяємо масив попереднього розміру на половини.</p>

<p>Оскільки n (кількість елементів) поділяється попередньо удвічі кожен раз, кількість ітерацій (кроків) пошуку буде логарифмічною функцією від
n, тобто O(logn). Таким чином, часова складність бінарного пошуку залежить від логарифма відносно кількості елементів у відсортованому масиві, що робить його дуже ефективним для великих наборів даних.</p>

<h3>-написати функцію, яка б могла повертати датафрейм з таким складом
полів: («n», «time»), де n – розмір масиву для функції linear_searh();</h3>

```python
import numpy as np
import time

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def create_data(max_n):
    results = [] 
    for n in range(1, max_n + 1):
        arr = np.arange(n)  
        target = np.random.randint(0, n)  
        start_time = time.time()
        linear_search(arr, target) 
        end_time = time.time()  
        execution_time = end_time - start_time 
        results.append([n, execution_time])
    return results

max_size = 100
data = create_data(max_size)

for entry in data:
    print(f"n: {entry[0]}, time: {entry[1]}")
```

<h3>− написати функцію, яка б могла повертати датафрейм з таким складом
полів: («n», «time»), де n – розмір масиву для функції bin_searh();</h3>

```python
import numpy as np
import time

def bin_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid  
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1  

def create_results(max_n):
    results = []
    for n in range(1, max_n + 1):
        arr = np.arange(n) 
        target = np.random.randint(0, n) 
        start_time = time.time() 
        bin_search(arr, target) 
        end_time = time.time()  
        execution_time = end_time - start_time  
        results.append((n, execution_time))

    return results

max_size = 100
results = create_results(max_size)

for result in results:
    n, execution_time = result
    print(f"n: {n}, time: {execution_time}")
```

Зазвичай бінарний пошук є більш ефективним (швидшим) за умови, що масив відсортований, оскільки він зменшує обсяг пошуку вдвічі на кожній ітерації.

Кожен з алгоритмів може бути більш ефективним:

Лінійний пошук:
<p>1.Часова складність: O(n), де n - розмір масиву.</p>
<p>2.Для лінійного пошуку час виконання зростає лінійно зі збільшенням розміру масивуn. Це означає, що чим більший масив, тим більше часу потрібно для пошуку.</p>

Бінарний пошук:
<p>1.Часова складність: O(logn), де n - розмір відсортованого масиву.</p>

Якщо масив відсортований і має великий розмір, рекомендується використовувати бінарний пошук для забезпечення ефективності пошуку.

<p>Шуканий об'єкт (елемент), який ми хочемо знайти в цій колекції.</p>
<h5>Вихідні дані:</h5>
<p>Визначення наявності чи відсутності шуканого об'єкта в колекції.</p>
<p>Якщо об'єкт знайдено, можливо, повернення його місцезнаходження (наприклад, індекс).</p>
<h5>Задача:</h5>
<p>Знайти шуканий об'єкт в заданій колекції даних.</p>
<h5>Умови:</h5>
<p>Вихід повинен бути коректним (відповідати наявності чи відсутності об'єкта в колекції).</p>
<p>Ефективність пошуку може залежати від розміру колекції та специфікацій задачі (наприклад, чи відсортована колекція).</p>
<h5>Методи пошуку:</h5>

<p>Бінарний пошук: застосовується до відсортованої колекції, розділяючи область пошуку попередньо.</p>
<h5>Оптимізація:</h5>
<p>Вибір оптимального методу пошуку залежить від типу даних і вимог до часу виконання.</p>
<p>У деяких випадках можна використовувати інші алгоритми для оптимізації пошуку (наприклад, хешування для швидкого доступу до даних).</p>

<h3>Які критерії можна використовувати для оцінювання ефективності
алгоритмів пошуку?</h3>

<h5>Існує кілька критеріїв, за допомогою яких можна оцінити ефективність алгоритмів пошуку. Нижче перераховано деякі з них:</h5>

<h5>Часова складність:</h5> Основний критерій ефективності алгоритму пошуку - це час, необхідний для виконання пошуку в залежності від розміру вхідних даних. Часова складність оцінює, як змінюється час виконання алгоритму при збільшенні обсягу вхідних даних. Зазвичай вимірюється у вигляді "O-нотації", такі як O(n) для лінійного пошуку або O(logn) для бінарного пошуку.
<h5>Пам'ятівідільність:</h5> Цей критерій оцінює обсяг пам'яті, необхідний для виконання алгоритму. Деякі алгоритми пошуку можуть вимагати більшої кількості пам'яті для зберігання даних або додаткових структур даних.

<h5>Індексованість даних:</h5> Цей критерій відноситься до наявності попередньо побудованих індексів або структур даних, що дозволяють прискорити пошук. Наприклад, в бінарному пошуку вимагається відсортований масив, що є формою індексу для швидкого пошуку.

<h5>Кількість порівнянь або операцій:</h5> Оцінка кількості порівнянь (або інших операцій) в алгоритмі дозволяє оцінити його продуктивність. Зазвичай швидкі алгоритми пошуку потребують менше порівнянь для досягнення результату.

<h5>Адаптивність до змін:</h5> Деякі алгоритми пошуку можуть бути більш адаптивними до змін у вхідних даних (наприклад, вставка нових елементів в масив) або до різних видів запитів.

<h5>Можливість розпаралелювання:</h5> Оцінка можливості алгоритму пошуку для розпаралелювання (використання багатьох обчислювальних ресурсів) може бути важливою в контексті великих обсягів даних.

<h3>Що таке лінійний пошук і яка його асимптотична складність?</h3>


Лінійний пошук - це простий алгоритм пошуку, який перевіряє кожен елемент колекції даних по порядку, щоб знайти шуканий елемент. 

<h5>Опис алгоритму лінійного пошуку:</h5>
<p>1.Починаємо з першого елементу колекції.</p>
<p>2.Порівнюємо поточний елемент з шуканим елементом.</p>
<p>3.Якщо поточний елемент співпадає з шуканим, пошук завершується успішно.</p>
<p>4.Якщо поточний елемент не співпадає з шуканим, переходимо до наступного елементу.</p>
<p>5.Продовжуємо цей процес до тих пір, поки не знайдемо шуканий елемент або не дійдемо до кінця колекції.</p>

Асимптотична складність лінійного пошуку визначається кількістю порівнянь, які потрібно зробити для знаходження шуканого елемента у колекції даних. При лінійному пошуку кожен елемент з колекції перевіряється рівно один раз у найгіршому випадку.

<h3>Що таке бінарний (двійковий) пошук і в чому полягає його головний
принцип роботи?</h3>

Бінарний пошук - це ефективний алгоритм пошуку елемента в відсортованій колекції даних. 

<h5>Принцип роботи бінарного пошуку:</h5>
<p>1.Почнемо з відсортованого масиву (або списку) елементів.</p>
<p>2.Визначимо дві границі: ліву границю (left) і праву границю (right) масиву. Ліва границя починається з першого елемента (індекс 0), а права границя - з останнього елемента (індекс n-1, де n - кількість елементів у масиві).</p>
<p>3.На кожній ітерації алгоритм обчислює середину (медіану) поточного підмасиву, яка є середнім індексом між left і right:
mid = (left + right) // 2</p>
<p>4.Порівнюємо значення елемента arr[mid] (елемент у середині поточного підмасиву) з шуканим значенням:
<p>4.1 Якщо arr[mid] дорівнює шуканому значенню, пошук завершується успішно і повертається індекс mid.</p>
<p>4.2 Якщо arr[mid] більше за шукане значення, змінюємо праву границю на mid - 1 для того, щоб шукати у лівій половині підмасиву.</p>
<p>4.3 Якщо arr[mid] менше за шукане значення, змінюємо ліву границю на mid + 1 для того, щоб шукати у правій половині підмасиву.</p></p>
<p>5.Повторюємо кроки 3-4, доки ліва границя left не стане більшою або рівною правій границі right. Якщо шукане значення знайдено, повертаємо його індекс; якщо не знайдено, повертаємо -1 (або інше значення, що відповідає відсутності шуканого елемента).</p>

<h5>Головні переваги бінарного пошуку:</h5>
<p>1.Ефективність: Часова складність бінарного пошуку становить O(logn), де n - кількість елементів у відсортованому масиві. Це означає, що час пошуку зменшується логарифмічно зі збільшенням розміру масиву, що робить бінарний пошук ефективним для великих колекцій даних.</p>
<p>2.Простота реалізації: Бінарний пошук є відносно простим у реалізації, алгоритмічно ефективним і чітко структурованим.</p>

<h3>Які переваги і недоліки використання бінарного пошуку порівняно з
лінійним?</h3>

<h5>Переваги використання бінарного пошуку порівняно з лінійним:</h5>
<p>Ефективність часу: Основною перевагою бінарного пошуку є його часова складність
O(log(n)), де 
n - кількість елементів у відсортованому масиві. Це означає, що час пошуку зменшується логарифмічно зі збільшенням розміру масиву. У порівнянні з лінійним пошуком 
O(n), бінарний пошук є набагато ефективнішим для великих обсягів даних.</p>
<p>Потреба у відсортованому масиві: Хоча відсортуваність масиву є вимогою для застосування бінарного пошуку, вона також може розглядатися як перевага. Відсортований масив легше підтримувати, а також може використовуватися для інших операцій, таких як бінарний пошук, що робить цей підхід більш гнучким у плануванні програм.</p>
<p>Простота реалізації: Бінарний пошук, хоча й вимагає відсортованого масиву, є відносно простим у реалізації. Цей алгоритм легше розуміти і імплементувати порівняно з більш складними алгоритмами пошуку.</p>

<h5>Недоліки використання бінарного пошуку порівняно з лінійним:</h5>

<p>Вимога до відсортованості масиву: Одним з основних обмежень бінарного пошуку є його вимога до відсортованості вхідного масиву. Якщо масив не відсортований, спершу потрібно виконати операцію сортування, що може зайняти додатковий час.</p>

<p>Неадаптивність до змін: Бінарний пошук не є оптимальним для ситуацій, коли масив часто змінюється або потребує частого вставлення нових елементів. Оскільки відсортованість масиву має підтримуватися для ефективного застосування бінарного пошуку, будь-які зміни у масиві можуть потребувати повторного сортування.</p>

<p>Невідповідність для деяких видів даних: Деякі дані можуть не підходити для відсортованого масиву або можуть бути складними для сортування, що ускладнює використання бінарного пошуку.</p>

<h3>Які ще алгоритми пошуку існують, окрім лінійного і бінарного?</h3>

<h5>Помимо лінійного і бінарного пошуку існує цілий ряд інших алгоритмів пошуку:</h5>

<h5>Інтерполяційний пошук (Interpolation Search):</h5>

<p>Цей алгоритм використовується для пошуку елемента у впорядкованих даних, де значення елементів рівномірно розподілені. Замість рівномірного розділення, як у бінарному пошуку, інтерполяційний пошук використовує формулу для оцінки положення шуканого елемента, що дозволяє ефективно скорочувати область пошуку. Часова складність в середньому становить 
O(log(log(n))), але вона може бути менш ефективною в неправильно розподілених даних.</p>
<h5>Фібоначчіевий пошук (Fibonacci Search):</h5>

<p>Цей алгоритм пошуку також використовує впорядковані дані, але замість ділення на пів у бінарному пошуку він використовує числа Фібоначчі для обчислення нової позиції пошуку. Часова складність такого пошуку також становить 

O(logn).</p>
<h5>Блочний пошук (Block Search):</h5>

<p>Цей алгоритм розділяє відсортований масив на блоки (частини) і потім використовує лінійний пошук в межах кожного блоку, щоб знайти відповідний блок. Після цього виконується бінарний пошук в межах знайденого блоку. Часова складність залежить від кількості блоків і способу поділу масиву.</p>
<h5>Хеш-таблиці (Hash Tables):</h5>

<p>Хеш-таблиці використовують хеш-функцію для перетворення ключа елемента в індекс таблиці, де він очікується знаходитися. Цей метод є дуже ефективним для пошуку, оскільки він забезпечує майже миттєвий доступ до елементів за допомогою хеш-функції. Проте він вимагає унікальних ключів і може мати колізії, які потребують додаткових операцій обробки.</p>
<h5>Дерева пошуку (Search Trees):</h5>

<p>Дерева пошуку, такі як бінарні дерева пошуку (Binary Search Trees) або червоно-чорні дерева (Red-Black Trees), також можуть використовуватися для ефективного пошуку. Вони забезпечують швидкий доступ до даних через оптимізовану структуру дерева, де кожен вузол має ключ і значення, і дані розділені на піддерева за значеннями ключів.</p>
<h5>Бойер-Мур пошук (Boyer-Moore Search):</h5>

<p>Це алгоритм пошуку для пошуку підстроки в тексті. Він використовує евристичні методи для прискорення пошуку шляхом використання інформації про останнє входження символів у шуканій підстроці.</p>
