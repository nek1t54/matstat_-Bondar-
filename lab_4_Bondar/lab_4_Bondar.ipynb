<h4>Реалізувати процедуру обчислення факторіалу за допомогою циклу for. Оцінити асимптотичну складніст 
алгоритму</h4>

```python
def factorial(n):
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i

    return result
```

<p>Асимптотична складність даного алгоритму:</p>

<p>1. У цьому алгоритмі ми використовуємо цикл for, який виконується n разів, де n - вхідний аргумент.</p>
<p>2. Кожна ітерація циклу має константний час виконання (основні операції множення та присвоєння).</p>
<p>3. Отже, асимптотична складність цього алгоритму є O(n), де n - число, для якого обчислюється факторіал.</p>

<h4>Оцінити асимптотичну складність
рекурсивного алгоритму обчислення факторіалу.</45>

<p>Алгоритм використовує рекурсію для обчислення факторіалу числа n.</p>
<h5>Розглянемо його асимптотичну складність:</h5>

<p>1.Рекурсивні виклики: Кожен виклик factorial(n) породжує ще один виклик factorial(n-1), і так далі, до базового випадку n == 0.</p>
<p>2.Глибина рекурсії: Глибина рекурсії в даному випадку буде n, оскільки ми робимо n рекурсивних викликів перед тим, як досягнемо базового випадку.</p>
<p>3.Часова складність: Кожен рекурсивний виклик factorial(k) для k від 1 до n виконується за константний час (операція множення). Отже, загальний час виконання алгоритму дорівнює сумі часів усіх рекурсивних викликів.</p>
<p>4.Асимптотична складність: Оскільки глибина рекурсії дорівнює n, а кожен рекурсивний виклик виконується за константний час, асимптотична складність рекурсивного алгоритму обчислення факторіалу буде O(n).</p>

<h4>Головна теорема рекурсії для сортування злиттям</h4>

Нехай T(n) позначає час виконання алгоритму сортування злиттям для масиву розміром n. Тоді за головною теоремою рекурсії для сортування злиттям маємо:

<h5>1. Базовий випадок:</h5> Якщо n = 1, сортування не потребує додаткових операцій, тобто T(1) = O(1).
<h5>2. Рекурентний випадок:</h5> При n > 1, алгоритм розділяє масив на дві частини розміром приблизно n/2 і рекурсивно сортує кожну з них. Після цього виконується операція злиття двох відсортованих підмасивів розміром приблизно n/2. Час на злиття двох підмасивів зазвичай є O(n).

<h3>У чому полягає ідея розпаралелювання обчислень і для чого вона використовується? Які з алгоритмів, наведених у цій лабораторній роботі, дозволяють можливість розпаралелювання?</h3>

<p>Основна мета розпаралелювання полягає у прискоренні виконання обчислень, зменшенні часу відповіді та підвищенні продуктивності обчислювальних систем. Розпаралелювання корисне тоді, коли задачу можна розділити на незалежні або слабо залежні підзадачі, які можуть бути вирішені паралельно. Це дозволяє використовувати більш ефективно обчислювальні ресурси, що забезпечує збільшення продуктивності та ефективності обчислень.</p>

<h5>Деякі з алгоритмів, які дозволяють можливість розпаралелювання, включають:</h5>

<p>1.Сортування злиттям (Merge Sort): Цей алгоритм може бути ефективно розпаралельованим, оскільки сортування кожної підмасиву може відбуватись одночасно, а потім відсортовані підмасиви можуть бути об'єднані.</p>
<p>2.Пошук максимуму у списку (Max Finding): Пошук максимального елементу у списку може бути розділений на пошук максимальних значень у підмасивах, що обробляються паралельно, а потім об'єднані для знаходження остаточного максимуму.</p>
<p>3.Обчислення чисел Фібоначчі (Fibonacci): Рекурсивні алгоритми для обчислення чисел Фібоначчі можуть бути розділені на більші менші завдання, які можуть бути виконані паралельно, наприклад, за допомогою паралельного виконання рекурсивних викликів.</p>
<p>4.Паралельне обчислення матричних операцій: Операції з матрицями, такі як множення матриць або обчислення детермінанту, можуть бути ефективно розпаралельовані, наприклад, за допомогою розбиття матриць на підблоки для паралельного обчислення.</p>

<h3>Які існують способи підвищення обчислювальної швидкості
алгоритмів? Який з них є найефективнішим?</h3>

<h5>Існує кілька способів підвищення обчислювальної швидкості алгоритмів. Ось декілька з них:</h5>

<p>Оптимізація алгоритмів: Першим кроком у підвищенні швидкості алгоритмів є їх оптимізація. Це включає вдосконалення алгоритмічної структури та уникання непотрібних операцій. Оптимізований алгоритм може виконуватись швидше за рахунок ефективнішого використання обчислювальних ресурсів.</p>

<p>Використання оптимізованих структур даних: Вибір правильної структури даних може значно вплинути на швидкість виконання алгоритму. Використання оптимізованих структур даних, таких як хеш-таблиці, бінарні дерева пошуку або кучі, може покращити продуктивність алгоритмів.</p>

<p>Паралельне програмування: Розпаралелювання обчислень дозволяє використовувати багатоядерні процесори та обчислювальні кластери для виконання різних частин алгоритму одночасно. Це дозволяє підвищити швидкість виконання, особливо для завдань, які можна розділити на незалежні підзадачі.</p>

<p>Використання спеціалізованих обчислювальних пристроїв: У деяких випадках використання спеціалізованих обчислювальних пристроїв, таких як графічні прискорювачі або тензорні процесори, може значно підвищити швидкість обчислень для конкретних типів завдань, наприклад, обробка графіки або нейронних мереж.</p>

<p>Кешування та оптимізація доступу до пам'яті: Ефективне використання кеш-пам'яті та оптимізація доступу до пам'яті може значно покращити швидкість виконання алгоритмів, особливо для алгоритмів, які мають велику кількість доступів до пам'яті.</p>
